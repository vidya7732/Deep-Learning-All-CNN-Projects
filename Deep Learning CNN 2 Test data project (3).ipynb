```python
path=r"C:\Users\vidya\Downloads\test"
```

```python
ct=["dogs","cats"]
ct.index("cats")
```

```python
Data=[]
for i in ct:
    print(os.path.join(path,i))
    A=os.path.join(path,i)
    for j in os.listdir(A):
        #print(j)
        #print(os.path.join(A,j))
        B=os.path.join(A,j)
        img=cv.imread(B)
        img1=cv.resize(img,(150,150))
        labels=ct.index(i)
        Data.append([img1,labels])
```

```python
F=[]
T=[]
for i,j in Data:
    F.append(i)
    T.append(j)
```

```python
F=np.array(F)
T=np.array(T)
```

```python
F.dtype
```

```python
F=F/255.
```

```python
F
```

```python
F.shape #557 images,150 height,150 width,3 channels RGB
```

```python
model=Sequential()
model.add(Conv2D(80,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2),strides=(1,1)))
model.add(Conv2D(60,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(40,(3,3),activation='relu'))
model.add(MaxPooling2D((2,2)))

model.add(Flatten())

model.add(Dense(50,input_shape=(150,150,3),activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(F,T,epochs=1,validation_split=0.15)
```
